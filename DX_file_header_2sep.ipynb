# Preparing Dataset

## Here we are taking y as dependent variable

```python
y = df.loc[:,"Malware"]
```

## Using train_test_split we are spliting our data into train and test format

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

## Now we are applying our ML Algo. XGBoost 

```python
import xgboost as xgb
xg = xgb.XGBClassifier(objective = "binary:logistic",random_state=42)
```

```python
xg.fit(X_train,y_train)
```

```python
predict=xg.predict(X_test)
```

## Here We are Checking accuracy of our model

```python
from sklearn.metrics import accuracy_score
```

```python
accur = accuracy_score(y_test,predict)
```

```python
print("Accuracy:-",accur)
```
